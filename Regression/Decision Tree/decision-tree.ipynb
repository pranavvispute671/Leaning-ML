```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv("../Polynomial Linear Regression/Position_Salaries.csv")
```

```python
x = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
```

```python
decision_regression = DecisionTreeRegressor(random_state=0)
decision_regression.fit(x,y)
```

```python
ypred = decision_regression.predict([[6.5],[7.55]])
```

```python
ypred
```

```python
Xgrid = np.arange(min(x), max(x), 0.1)
Xgrid = Xgrid.reshape((len(Xgrid),1))
plt.scatter(x,y, color ="red")
plt.plot(Xgrid, decision_regression.predict(Xgrid), color="blue")
plt.title('Decision tree Regression')
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
