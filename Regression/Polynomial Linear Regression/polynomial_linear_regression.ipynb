```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

LINEAR REGRESSION

```python
data = pd.read_csv("Position_Salaries.csv")
x = data.iloc[:,1:-1]
y = data.iloc[:,-1]
```

```python
linear_regressor = LinearRegression()
linear_regressor.fit(x, y)
```

POLYNOMIAL LINEAR REGRESSION

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_transformer = PolynomialFeatures(degree=3)
poly_x = poly_transformer.fit_transform(x)
```

```python
poly_regressor = LinearRegression()
poly_regressor.fit(poly_x, y)
```

Visualizing Linear Regression Results

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, linear_regressor.predict(x), color = "blue" )
plt.title("Linear Regression results")
plt.xlabel("Position")
plt.ylabel("Salary Expectation")
plt.show()
```

Visualizing Polynomial Linear Regression Results

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, poly_regressor.predict(poly_x), color = "blue")
plt.plot(x, linear_regressor.predict(x), color = "orange" )
plt.title("Polynomial Linear Regression results")
plt.xlabel("Position")
plt.ylabel("Salary Expectation")
plt.show()
```

Predicting new result from Linear Regression

```python
y_linear = linear_regressor.predict([[6.5]])
y_linear
```

```python
y_poly = poly_regressor.predict(poly_transformer.fit_transform([[6.5]]))
y_poly
```
