```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("Ice_cream selling data.csv")
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

Polynomial Linear Regression. This should work

```python
poly_trans = PolynomialFeatures(degree=2)
```

```python
polyX = poly_trans.fit_transform(x)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(polyX, y, test_size=0.3, random_state=42)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(xtrain, ytrain)
```

```python
# Sort x for smooth curve plotting
x_sorted = np.sort(x.values.flatten()).reshape(-1, 1)
x_poly_sorted = poly_trans.transform(x_sorted)

# Predict on the sorted x values
y_pred_sorted = lin_reg.predict(x_poly_sorted)

# Plot original data
plt.scatter(x, y, color='red')

# Plot regression curve
plt.plot(x_sorted, y_pred_sorted, color='blue')
plt.xlabel("Temperature")
plt.ylabel("Ice Cream Sales")
plt.title("Polynomial Regression")
plt.show()
```
